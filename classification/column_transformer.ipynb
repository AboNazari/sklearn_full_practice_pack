```python
import pandas as pd 
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
from sklearn.compose import make_column_transformer
from sklearn.datasets import fetch_openml
from sklearn.impute import SimpleImputer
```

```python
titanic, y = fetch_openml(name="titanic", version=1, return_X_y=True)
titanic
```

```python
ohe = OneHotEncoder(sparse_output=False)
ode = OrdinalEncoder()
```

```python
mean_imp = SimpleImputer(strategy="mean")
const_imp = SimpleImputer(strategy='constant', fill_value="M")
```

```python
ct1 = make_column_transformer(
    (const_imp, ['cabin']),
    remainder="passthrough"
)
ct1.set_output(transform="pandas")
titanic = ct1.fit_transform(titanic)
titanic.rename(columns={"remainder__sex":"sex", "simpleimputer__cabin":"cabin"},inplace=True)
```

```python
ct = make_column_transformer(
    (ohe, ['sex']),
    (ode, ['cabin']),
    remainder="passthrough"
)
```

```python
ct.set_output(transform="pandas")
```

```python
tr_titanic = ct.fit_transform(titanic)
tr_titanic
```
